```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10,1000)
```

```python
plt.plot(x,np.sin(x))  # first x axis and then y-axis value
plt.plot(x,np.cos(x))  #.plot for line plot

#plt.show()    #this is use for other paltform jupyter islye best ha ye phr b show kr dy ga
```

```python
#SUBPLOT
#matplotlib styles
plt.subplot(2,1,1)  # 2 rows first column me  1_first row me sin ka plot kr dy
plt.plot(x,np.sin(x))   #plt.subplot is state based programming

plt.subplot(2,1,2)  # 2 rows first column me  second row me cos ka plot kr dy
plt.plot(x,np.cos(x))
```

```python
fig,ax = plt.subplots(2)  # 2 things return krta subplots function 1 figure ka default size wo fig me save hojata
# and axis jo grah k wo ax me save hn gyb jese yaha 2 axis ha ax[0],ax[1]

ax[0].plot(x,np.sin(x))
ax[1].plot(x,np.cos(x))
```

```python
plt.style.use("seaborn-whitegrid")
#hashcode use for colors that is better  #000 for black
x=np.linspace(0,10,1000)

plt.plot(x,np.sin(x),color="b",linestyle="solid",label="sin(x)")   # color which color to use for graph line
plt.plot(x,np.sin(x-1),color="g",linestyle="dashed",label="sin(x-1)")  # linestyle which t=style use for line
plt.plot(x,np.sin(x-2),color="r",linestyle="dotted",label="sin(x-2)")#label use for show in graph which line show for which plot
plt.plot(x,np.sin(x-3),color="#000",linestyle="dashdot",label="sin(x-3)")
plt.plot(x,np.sin(x-5),":b",label="sin(x-5)")
plt.plot(x,np.sin(x-4),color="chartreuse",linestyle="-.",label="sin(x-4)")

plt.title("Sin Graph")
plt.xlabel("x")  # x label py jo value show kraani ha
plt.ylabel("sinx")  # y label py jo values show krani ha
plt.legend()

plt.show()
```
