```python
import matplotlib.pyplot as plt  #for visualization
import numpy as np  #for array
import pandas as pd  #for data read and cleaning
import seaborn as sns  #for EDA
```

```python
x=np.linspace(0,10,1000)
```

```python
plt.plot(x,np.sin(x))  # first x axis and then y-axis value
plt.plot(x,np.cos(x))  #.plot for line plot

#plt.show()    #this is use for other paltform jupyter islye best ha ye phr b show kr dy ga
```

```python
#SUBPLOT
#matplotlib styles
plt.subplot(2,1,1)  # 2 rows first column me  1_first row me sin ka plot kr dy
plt.plot(x,np.sin(x))   #plt.subplot is state based programming

plt.subplot(2,1,2)  # 2 rows first column me  second row me cos ka plot kr dy
plt.plot(x,np.cos(x))
```

```python
fig,ax = plt.subplots(2)  # 2 things return krta subplots function 1 figure ka default size wo fig me save hojata
# and axis jo graph k wo ax me save hn gyb jese yaha 2 axis ha ax[0],ax[1]

ax[0].plot(x,np.sin(x))  #ax[0] means ye first graph me plot ho
ax[1].plot(x,np.cos(x))  #ax[1] means plot in second graph
```

```python
plt.style.use("seaborn-whitegrid") #for graph grid color layout
#hashcode use for colors that is better  #000 for black
x=np.linspace(0,10,1000)

plt.plot(x,np.sin(x),color="b",linestyle="solid",label="sin(x)")   # color which color to use for graph line
plt.plot(x,np.sin(x-1),color="g",linestyle="dashed",label="sin(x-1)")  # linestyle which t=style use for line
plt.plot(x,np.sin(x-2),color="r",linestyle="dotted",label="sin(x-2)")#label use for show in graph which line show for which plot
plt.plot(x,np.sin(x-3),color="#000",linestyle="dashdot",label="sin(x-3)")
plt.plot(x,np.sin(x-5),":b",label="sin(x-5)")
plt.plot(x,np.sin(x-4),color="chartreuse",linestyle="-.",label="sin(x-4)")

plt.title("Sin Graph")
plt.xlabel("x")  # x label py jo value show kraani ha
plt.ylabel("sinx")  # y label py jo values show krani ha
plt.legend()

plt.show()
```

```python
x2= np.random.randn(100)  #generate 100 random values range -1 to 1
y2= np.random.randn(100)
colors = np.random.rand(100)  #rand generate 100 values range 0 to 1
size = 1000*np.random.rand(100) 
```

```python
#in scatter plot c use for colors and s for sizes and alpha for lightness or transperancy
plt.subplot(1,2,1)
plt.scatter(x2,y2,c=colors,s=size) 
plt.colorbar()
plt.subplot(1,2,2)
plt.scatter(x2,y2,c=colors,s=size,alpha=0.5)  #show all points behind any points
plt.colorbar() #show color bar in graph
```

```python
#in scatter plot c use for colors and s for sizes and alpha for lightness or transperancy
plt.subplot(1,2,1)
plt.scatter(x2,y2,c=colors,s=size) 
plt.colorbar()
plt.subplot(1,2,2)
plt.scatter(x2,y2,c=colors,s=size,alpha=0.5)  #show all points behind any points
plt.colorbar() #show color bar in graph
plt.tight_layout() #create space between graph
```

# Histogram

```python
data = np.random.randn(1000)
plt.hist(data)
```

```python
plt.subplot(1,3,1)
plt.hist(data,bins=30,color='red',alpha=0.5)   #bins means kitna spread or zoom out kr k show krana
plt.subplot(1,3,2)
plt.hist(data,bins=80,color='steelblue',histtype='stepfilled') #histype used for which type of graph to draw
plt.subplot(1,3,3)
plt.hist(data,bins=80,color='red',histtype='step') 
plt.tight_layout(pad=1.8)
```

```python
z1 = np.random.normal(0,0.8,2000)
z2 = np.random.normal(-2,0.5,1000)
z3 = np.random.normal(3,2,1000)
dict_extra = dict(histtype='stepfilled',bins=10,alpha=0.3)  #this is dictionary
#histtype is key and stepfilled is the value

plt.hist(z1,**dict_extra) #double asteric used when pass dictionary to the hist function but if we pass list then use single asteric
plt.hist(z2,**dict_extra) 
plt.hist(z3,**dict_extra)
```

# Ledends

```python
x=np.linspace(0,10,1000)
plt.subplot(3,1,1)
plt.plot(x,np.sin(x),'-b',label='sinx')
plt.plot(x,np.cos(x),'--r',label='cosx')
plt.legend() #by default position set of label

plt.subplot(3,1,2)
plt.plot(x,np.sin(x),'-b',label='sinx')
plt.plot(x,np.cos(x),'--r',label='cosx')
plt.legend(loc='center') #set position of label using loc keyword and assign value 
#either use words or either use number from 0 to 10

plt.subplot(3,1,3)
plt.plot(x,np.sin(x),'-b',label='sinx')
plt.plot(x,np.cos(x),'--r',label='cosx')
plt.legend(loc=10,ncol=2)  #ncol use for divide labels into two columns
```

```python
x,y = np.random.multivariate_normal([0,0],[[1,1],[1,2]],10000).T
plt.subplot(2,1,1)
plt.hist2d(x,y,bins=20)
plt.subplot(2,1,2)
plt.hist2d(x,y,bins=20,cmap='Blues')
plt.show()
```
