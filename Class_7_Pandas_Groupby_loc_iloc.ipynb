### add a num_of_cast column   ASSIGNMENT

```python
data.head()
```

```python
data["cast"] = data['cast'].replace(0, "None")
data
```

```python
def set_cast(val):
    if val == 'None':
        return 0
    return len(val.split(', '))
```

```python
data['num_of_cast'] = data['cast'].apply(set_cast)
data
```
