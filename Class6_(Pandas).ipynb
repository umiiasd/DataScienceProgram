### ADD COLUMN
#### function of pandas-->apply

```python
data.shape  #in our data, 800 rows and 13 columns
```

```python
data["Attack"]
```

```python
data["Attack"].mean()  #mean of attack column 
```

```python
#we want to add column attack_high_low --> Attack High , Attack Low 
#when value of attck is greater then 79 print Attack High otherwise print Attack low

attack_mean = data["Attack"].mean()

def set_attack(val):   #set_attack is function to take value of attack in this column
    if val>attack_mean:
        return "Attack High"
    elif val==attack_mean:
        return "Attack Neutral"
    else:
        return "Attack Low"
```

```python
#now we add column  #apply is function of pandas that apply userdefined function to datasets
data["Attack_High_Low"]=data["Attack"].apply(set_attack)  #get all values of attack one by one and send to set_attack method then check condition
data
```

```python
data.loc[(data.Attack_High_Low=="Attack High"),["Name","Attack","Attack_High_Low"]] 
```

```python
data.loc[(data.Attack_High_Low=="Attack Neutral"),["Name","Attack","Attack_High_Low"]]
```

```python
#ADD Column of Speed_High_Low and HP_High_Low
Speed_mean = data["Speed"].mean()
print("Mean of Speed = ",Speed_mean)

HP_mean = data["HP"].mean()
print("Mean of HP = ",HP_mean)

def set_speed(val):
    if val>Speed_mean:
        return "Speed High"
    elif val==Speed_mean:
        return "Speed Neutral"
    else:
        return "Speed Low"
    
def set_hp(val):
    if val>HP_mean:
        return "HP High"
    elif val==HP_mean:
        return "HP Neutral"
    else:
        return "HP Low"
```

```python
data["Speed_High_Low"]=data["Speed"].apply(set_speed)
data["HP_High_Low"]=data["HP"].apply(set_hp)
data
```
