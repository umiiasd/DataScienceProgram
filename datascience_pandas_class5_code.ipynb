# PANDAS FUNCTION

LOC  #for label based selection
ILOC #for index based selection

```python
data.loc[(data.rating=="PG-13") | (data["release year"]==2011)]  #use or operator means anyone from both
```

```python
data.loc[(data.rating=="PG-13") & (data["release year"]==2011)]  #use and operator means both true
```

```python
data.isnull().sum()  #get info about dataset which which value is null
```

```python
data.notnull().sum()  #which column has no null values
```

```python
data.loc[(data.rating=="TV-14")]  #rating is TV-14
```

```python
data.loc[(data.rating=="TV-14") & (data["user rating score"]>=90)]  #rating is TV-14 and rating score is 90 or above
```

```python
data.loc[(data.rating=="TV-14"),["title","ratingLevel","ratingDescription"]] #in pandas first is row  and second is column-->separate by comma of row/column
```

```python
data.loc[(data.rating=="TV-14"),["title","ratingLevel","ratingDescription"]].isnull().sum() #how many null values in new data
```

```python
data.loc[(data.rating=="TV-14"),["title","ratingLevel","ratingDescription"]].shape
```
