```python
#Question#1
"""Create a new column in the provided dataset. Column name should be num_of_cast. 
The assignment was already explained in class. You have to append or add a column in the dataset 
which contains the total number of cast in that particular movie or Tv-Show."""
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("datasets/netflix_data.csv")
```

```python
data
```

```python
data["cast"] = data['cast'].replace(np.nan, "None")
data.head()
```

```python
def set_cast(val):
    if val == 'None':
        return 0
    else:
        return len(val.split(', '))
```

```python
data['num_of_cast'] = data['cast'].apply(set_cast)
data
```
